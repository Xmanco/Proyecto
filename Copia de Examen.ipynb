```
    def g1(x):
      return (  (-0.9*(x**2)) + (1.7*x) + 2.5 )
    def gg1(x):
      return ( (-1.8*x) + 1.7 )
    def ffg2(x):
      return ( -2.5/(-0.9*x + 1.7) )
```

```
import matplotlib.pyplot as plt
import math
def punto_fijo():
  print("xinicial = 2")
  fg1p()

def fg1p():
  x_vec=[]
  x_vec.append(2)
  contador = 0
  error_vec=[]
  error_vec.append(1)
  criterio_paro = 0.0001
  iteraciones = True
  while(error_vec[contador] > criterio_paro):
    x_temp = ffg2(x_vec[contador])
    x_vec.append(x_temp)
    e_temp = abs(x_vec[contador+1]-x_vec[contador])
    error_vec.append(e_temp)
    contador += 1
    if (x_vec[contador]>=2000):
      iteraciones = False
      break
  print(iteraciones)
  print("Raiz "+str(x_vec[contador-1]))
  plt.plot(x_vec, label = "x_r")
  plt.plot(error_vec, label = "error")
  plt.legend()
  plt.show()
punto_fijo()
```

```
import matplotlib.pyplot as plt
import math
def newton():
  print("Metodo Newton Raphson")
  print("x0 = 2")
  fg2()

def fg2():
  x_vec=[]
  x_vec.append(2)
  contador = 0
  error_vec=[]
  error_vec.append(1)
  criterio_paro = 0.0001
  iteraciones = True
  while(error_vec[contador] > criterio_paro):
    xx = g1(x_vec[contador])
    yy = gg1(x_vec[contador])
    div = (xx / yy) if yy != 0 else 0
    x_temp = (x_vec[contador]-div)
    x_vec.append(x_temp)
    e_temp = abs(x_vec[contador+1]-x_vec[contador])
    error_vec.append(e_temp)
    contador += 1
    if (x_vec[contador]>=2000):
      iteraciones = False
      break
  print("Raiz "+str(x_vec[contador-1]))
  print(iteraciones)
  plt.plot(x_vec, label = "x_r")
  plt.plot(error_vec, label = "error")
  plt.legend()
  plt.show()
newton()
```
